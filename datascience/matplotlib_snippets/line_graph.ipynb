```python
x = [-3, -2, -1, 0, 1, 2, 3]
y = [-2.5, -1.5, -0.5, 0.5, 1.5, 2.5, 3.5]
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Straight Line Plot')
plt.show()
```
